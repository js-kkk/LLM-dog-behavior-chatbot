```python
import os
import whisper

ffmpeg_path = "C:\\prg\\ffmpeg\\ffmpeg-master-latest-win64-gpl-shared\\bin"

AUDIO_DIR = "./data/Audio"  # ì˜¤ë””ì˜¤ ì¶”ì¶œ ì €ì¥ ë”•ë ‰í† ë¦¬
TEXT_DIR = "./data/Text"    # ìë§‰ ì €ì¥ ë””ë ‰í† ë¦¬

# ë””ë ‰í† ë¦¬ í™•ì¸í›„ ì—†ìœ¼ë©´ ìƒì„±ì„±
os.makedirs(TEXT_DIR, exist_ok=True)

# ffmpeg ê²½ë¡œ ëª…ì‹œ
os.environ["PATH"] += os.pathsep + rf"{ffmpeg_path}"

# whisper ëª¨ë¸ ì„¤ì •
model = whisper.load_model("small").to("cuda")

for file in os.listdir(AUDIO_DIR):
    if file.endswith(".mp3"):
        audio_path = os.path.join(AUDIO_DIR, file)
        output = os.path.join(TEXT_DIR, file.replace(".mp3",".txt"))
        
        # ì¤‘ë³µ ì—¬ë¶€ í™•ì¸
        if os.path.exists(output):
            print(f"ì´ë¯¸ ì¡´ì¬í•¨, ê±´ë„ˆëœ€: {file}")
            continue

        print(f"ì²˜ë¦¬ ì‹œì‘: {file}")

        try:
            # ì˜¤ë””ì˜¤ -> í…ìŠ¤íŠ¸ ì²˜ë¦¬
            result = model.transcribe(audio_path, language="ko", fp16=True)
            text = result["text"]

            
            # ìë§‰ TEXT ì €ì¥
            with open(output, 'w', encoding='utf-8') as f:
                f.write(text)

            print(f"ì²˜ë¦¬ ì™„ë£Œ: {file}")
        
        except Exception as e:
            print(f"ì—ëŸ¬ : {file}")
```

```python
from llm.call_llm import call_llm
import json
import re

def create_json_with_llm(dog_text):
    system_prompt=""
    user_prompt=""

    system_prompt = f'''
    ë‹¹ì‹ ì€ ë°˜ë ¤ê²¬ ì „ë¬¸ê°€ì…ë‹ˆë‹¤. ì‚¬ìš©ìê°€ ì…ë ¥í•œ ê²¬ì¢… ì„¤ëª… í…ìŠ¤íŠ¸ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ë‹¤ìŒ JSON êµ¬ì¡°ë¥¼ ìƒì„±í•˜ì„¸ìš”:

    1. text_chunks  = ì˜ë¯¸ ë‹¨ìœ„ë¡œ 300~500ì ì´ë‚´ë¡œ ë¶„í• 
    2. id = breed_(ê²¬ì¢… ì˜ì–´ ì´ë¦„)
    3. title = (ê²½ì¢… í•œêµ­ ì´ë¦„) ê²¬ì¢… ì •ë³´
    4. metadataì˜ ê²°ê³¼ê°’ì€ ì˜ì–´ë¡œë§Œ ì¶œë ¥
    5. ì˜¤íƒ€ê°™ê±°ë‚˜ ë¬¸ì œê°€ ë˜ëŠ” í…ìŠ¤íŠ¸ëŠ” í–‰ë‹¹ ë¬¸ë§¥ì— ë§ê²Œ ìˆ˜ì •
    5. ë‹¤ìŒ í˜•ì‹ì„ JSONìœ¼ë¡œë§Œ ì •í™•íˆ ì¶œë ¥í•˜ì„¸ìš”. ì„¤ëª… ê¸ˆì§€
    ë°˜í™˜ í˜•ì‹:
    {{
      "id": "...",
      "title": "...",
      "text_chunks": [...],
      "metadata": {{
        "origin": "...",
        "size": "...",
        "breed_group": "...",
        "fur_type": "...",
        "trainability": "...",
        "activity_level": "...",
        "separation_anxiety": true,
        "family_friendly": true,
        "aggression_tendency": "...",
        "health_issues": [...],
        "grooming_needs": "...",
        "guard_ability": "...",
        "difficulty_score": ...
      }}
    }}
    '''

    user_prompt = f'''
    ì•„ë˜ëŠ” í•´ë‹¹ ê²¬ì¢…ì— ëŒ€í•œ ì„¤ëª…ì…ë‹ˆë‹¤:
    {dog_text}
    '''

    response = call_llm(system_prompt, user_prompt)
    
    try:
        return json.loads(response)
    except json.JSONDecodeError:
        match = re.search(r'\{.*\}', response, re.DOTALL)
        if match:
            return json.loads(match.group())
        else:
            raise ValueError("GPT ì‘ë‹µì—ì„œ JSONì„ ì¶”ì¶œí•˜ì§€ ëª»í–ˆìŠµë‹ˆë‹¤.")
```

```python
import os
import json

TEXT_DIR = "./data/Text"
JSON_DIR = "./data/Json"
os.makedirs(JSON_DIR, exist_ok=True)

for file in os.listdir(TEXT_DIR):
    if file.endswith(".txt"):
        # íŒŒì¼ ê²½ë¡œ ì„¤ì •
        file_path = os.path.join(TEXT_DIR, file)
        # ì½ê¸° í…ìŠ¤íŠ¸ ë‚´ìš© ê°€ì ¸ì˜´
        with open(file_path, "r", encoding="utf-8") as f:
            dog_text = f.read()

        try:
            print(f"ì²˜ë¦¬ ì‹œì‘ : {file}")
            # LLMì„ ì´ìš©í•œ json ê²°ê³¼ ê°’ ìƒì„±
            result = create_json_with_llm(dog_text)
            # ì €ì¥ ê²½ë¡œ, ì´ë¦„ ë³€ê²½
            output_path = os.path.join(JSON_DIR, f"{result['title']}.json")
            # ì“°ê¸° json íŒŒì‹±í›„ ì €ì¥
            with open(output_path, "w", encoding="utf-8") as out:
                json.dump(result, out, ensure_ascii=False, indent=2)
            print(f"ìƒì„± ì™„ë£Œ: {output_path}")
        except Exception as e:
            print(f"ì²˜ë¦¬ ì‹¤íŒ¨: {file}")
```

```python
import chromadb
from sentence_transformers import SentenceTransformer
import json

JSON_DIR = "./data/Json"

# Chromadb
client = chromadb.PersistentClient(path='./chroma_db')
# collection = client.create_collection(name = "dog_breed")
collection = client.get_collection(name= "dog_breed")

# embedding ëª¨ë¸ ë¡œë“œ
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

def normalize_metadata(metadata: dict) -> dict:
    return {
        k: v if isinstance(v, (str, int, float, bool)) else str(v)
        for k, v in metadata.items()
    }

# json íŒŒì¼ ë¡œë“œ í›„ ì…ë ¥
for file in os.listdir(JSON_DIR):
    if file.endswith(".json"):
        file_path = os.path.join(JSON_DIR, file)

        with open(file_path, "r", encoding="utf-8") as f:
            data = json.load(f)

        id = data["id"]
        title = data["title"]
        chunks = data["text_chunks"]
        metadata = normalize_metadata(data["metadata"])

        print(f"ë°ì´í„° ì£¼ì… ì‹œì‘ : {file}")

        # Chromadbì— ë°ì´í„° ì…ë ¥, ì„ë² ë”©
        for i, chunk in enumerate(chunks):
            embedding = embedding_model.encode([chunk]).tolist()[0]

            collection.add(
                ids=[f"{id}_chunk_{i}"],
                documents=[chunk],
                metadatas=[{
                    **metadata,
                    "title": title,
                    "chunk_index": i
                }],
                embeddings=[embedding]
            )

        print(f"ë°ì´í„° ì£¼ì… ì™„ë£Œ: {file}")


```

```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from llm.call_llm import call_llm
from sentence_transformers import SentenceTransformer

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# ì§ˆë¬¸ ì…ë ¥
query = input("ì§ˆë¬¸ì„ ì…ë ¥: ")

# ì„ë² ë”© ëª¨ë¸ ë¡œë“œ
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# ì§ˆë¬¸ -> ì„ë² ë”© ë²¡í„° ë³€í™˜
query_vector = embedding_model.encode([query]).tolist()[0]

# Chroma ê²€ìƒ‰
results = collection.query(
    query_embeddings=[query_vector],
    n_results=3,
    include=["documents", "metadatas"]
)

# ê²€ìƒ‰ëœ í…ìŠ¤íŠ¸ ì¡°í•©
retrieved_text = "\n\n".join(results["documents"][0])

print(f"ìœ ì‚¬ ë¬¸êµ¬ : {retrieved_text}")

# í”„ë¡¬í”„íŠ¸ ìƒì„±
system_prompt = "ì½”ê¸° ê²¬ì¢… ì •ë³´"
user_prompt = f"""
[ì§ˆë¬¸]
{query}

[ê²¬ì¢… ì •ë³´]
{retrieved_text}
"""

# LLM
response = call_llm(system_prompt, user_prompt)

# ì¶œë ¥
print(f"\në‹µë³€: {response}")
```
